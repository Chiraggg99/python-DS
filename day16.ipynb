```python
df = pd.read_csv('Used_Bikes.csv')
df.head()
```

```python
df.drop_duplicates(inplace=True) # 25k
```

```python
df['brand'].value_counts()
```

```python
brand_name = list(df['brand'].value_counts().keys())
brand_name
```

```python
no_of_counts = list(df['brand'].value_counts().values)
no_of_counts
```

```python
plt.bar(brand_name, no_of_counts)
plt.show()
```

```python
plt.figure(figsize=(10,4)) ## (width, height)
brand_name = list(df['brand'].value_counts().keys())
no_of_counts = list(df['brand'].value_counts().values)
plt.xticks(rotation=90)
plt.yticks(rotation=60)
plt.bar(brand_name, no_of_counts,color='red',ec = "black")
plt.savefig('brandpopulation')
plt.show()
## ec= edge color
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['owner'].value_counts()
```

```python
plt.figure(figsize=(10,6)) ## (width, height)
owner_type = list(df['owner'].value_counts().keys())
no_of_owner = list(df['owner'].value_counts().values)
plt.xticks(rotation=90)
plt.yticks(rotation=60)
plt.bar(owner_type, no_of_owner,color='aqua',ec = "black",lw=3)
plt.savefig('brandpopulation')
plt.show()
```

```python
rng = np.random.RandomState(0)
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(rng.rand(5), rng.rand(5), marker,
             label="marker='{0}'".format(marker))
plt.legend(numpoints=1)
plt.xlim(0, 1.8);
```
