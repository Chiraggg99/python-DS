```python
brand_group = df.groupby('brand')
brand_group
```

```python
brand_group.get_group('TVS')
```

```python
brand_group[['price']].min()
## minimum value of every bike 
```

```python
brand_group[['price']].max()
```

```python
brand_group[['price']].mean()
```

```python
brand_group['price'].agg(min_price='min',max_price='max',avg_price='mean')
```

```python
brand_group['kms_driven'].agg(min_driven='min',max_driven='max',avg_driven='mean')
```

```python
df['owner'].value_counts()
```

```python
# owner,min,max,average
# price,kms_driven

brand_group['price'].agg(min_price='min',max_price='max',avg_price='mean',min_driven='min',max_driven='max',avg_driven='mean')
```
